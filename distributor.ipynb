금액의 결측치를 수정

```python
uriage_data.isnull().any(axis=0)
```

```python
price_check_null = uriage_data["item_price"].isnull()
# check_null
price_check_null
```

```python
for li in list(uriage_data.loc[price_check_null, "item_name"].unique()):  
    #### 결측치가 있는 상품과 같은 상품중 정상값인 금액
    price = uriage_data.loc[(~price_check_null) & (uriage_data["item_name"] == li), "item_price"].max()
    #### 결측치가 있는 상품과 같은 상품중 null 값 데이터 추출
    uriage_data["item_price"].loc[(price_check_null) & (uriage_data["item_name"]==li)] = price
    
```

<font color=pink>날짜 데이터 수정(to_timedelta, to_datetime)</font>


```python
is_date_error = kokyaku_data["등록일"].astype("str").str.isdigit()
is_date_error.sum()
```

<font color =red>to_timedelta 날짜를 하루 단위로 변경후 to_datetime("1900/01/01")로 연월일로 변환</font>


```python
fromSerial = pd.to_timedelta(kokyaku_data.loc[is_date_error, "등록일"].astype("float") , unit = "D") + pd.to_datetime("1900/01/01")
```

```python
fromString = pd.to_datetime(kokyaku_data.loc[~is_date_error, "등록일"])
fromString
```

```python
kokyaku_data["등록일"] = pd.concat([fromSerial,fromString])
kokyaku_data["등록일"].tail()
```

```python
kokyaku_data["등록연월"] = kokyaku_data["등록일"].dt.strftime("%Y%m") 
```

```python
result = kokyaku_data.groupby("등록연월").count()["고객이름"]
print(result)
print(len(kokyaku_data))
```
